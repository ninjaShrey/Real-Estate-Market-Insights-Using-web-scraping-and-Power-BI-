```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import time
import json

chromedriver_path = r"C:\Users\shreya paul\OneDrive\Desktop\Jupyter projects\chromedriver.exe"
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service)
URL = "https://www.magicbricks.com/paying-guest-for-girls-in-kolkata-pppfr"
driver.get(URL)
time.sleep(7)

soup = BeautifulSoup(driver.page_source, "html.parser")

pg_listings = soup.find_all('div', class_='m-srp-card')
print(f"Found {len(pg_listings)} PG listings on the page")

titles, locations, genders = [], [], []
single_rents, twin_rents, triple_rents = [], [], []
ratings, food_charges, beds_availabilities = [], [], []

for pg in pg_listings:
    # Title
    meta_name = pg.find('meta', itemprop='name')
    title = meta_name['content'].strip() if meta_name else 'N/A'
    # Location
    meta_location = pg.find('meta', itemprop='addressLocality')
    location = meta_location['content'].strip() if meta_location else 'N/A'
    
    # Gender & Sharing
    span_hidden = pg.find('span', class_='hidden')
    gender = 'N/A'
    single_rent = 'N/A'
    twin_rent = 'N/A'
    triple_rent = 'N/A'
    if span_hidden:
        gender = span_hidden.get('data-pgfor', 'N/A')
        avail_json = span_hidden.get('data-avail', '[]')
        try:
            avail_data = json.loads(avail_json)
            for room in avail_data:
                sharing, price = list(room.values())[0].split(",")
                sharing_lower = sharing.lower()
                if 'single' in sharing_lower:
                    single_rent = price.strip()
                elif 'twin' in sharing_lower:
                    twin_rent = price.strip()
                elif 'triple' in sharing_lower:
                    triple_rent = price.strip()
        except Exception as e:
            print(f"Error parsing avail JSON: {e}")

    # Rating
    rating_div = pg.find('div', class_='pg-rating__tag--text1')
    rating = 'N/A'
    if rating_div:
        rating = rating_div.get_text(strip=True).split('/')[0]
        
    food_status = 'Not Included'
    food_span = pg.find('span', class_='m-srp-card__for__item')
    if food_span and 'Food' in food_span.get_text(strip=True):
        food_status = 'Included'

    beds_availability = 'Not Available'
    beds_div = pg.find('div', class_='m-srp-card__beds-available')
    if beds_div:
        beds_availability = beds_div.get_text(strip=True)
        
    titles.append(title)
    locations.append(location)
    genders.append(gender)
    single_rents.append(single_rent)
    twin_rents.append(twin_rent)
    triple_rents.append(triple_rent)
    ratings.append(rating)
    food_charges.append(food_status)
    beds_availabilities.append(beds_availability)

df = pd.DataFrame({
    'Title': titles,
    'Location': locations,
    'Gender': genders,
    'Single Sharing Rent': single_rents,
    'Twin Sharing Rent': twin_rents,
    'Triple Sharing Rent': triple_rents,
    'Rating': ratings,
    'Food Charges': food_charges,
    'Beds Availability': beds_availabilities
})

df.to_csv('kolkata_pg_listings_full.csv', index=False)
print("CSV 'kolkata_pg_listings_full.csv' created successfully!")

driver.quit()
```
